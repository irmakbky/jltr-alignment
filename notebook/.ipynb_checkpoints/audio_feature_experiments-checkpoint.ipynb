```python
from mus_align import mesa13, smr
from mus_align.eval import evaluate
```

```python
import os
import pickle
import numpy as np
```

```python
! pip install tqdm
from tqdm import tqdm
```

```python
tags = ["andante_con_moto", "lux_aeterna", "maple_leaf_rag", "pink_chinese_rag", "prelude_in_c", "the_maid",
       "omaggio_a_brahms", "promenade", "a_lesson", "pastorale", "berceuse", "gymnopedie", "liebestraum"]
```

```python
representation_accs_mesa13 = {k: 0 for k in ['frame_preds', 'frames', 'onset_preds', 'midi']}
for tag in tqdm(tags):
    for representation in ['frame_preds', 'frames', 'onset_preds', 'midi']:
        aligned = mesa13.load_audio_feature_alignment(tag=tag, audio_feature=representation)
        gt = mesa13.load_mesa13_alignment(tag=tag)
        acc = round(evaluate(pred=aligned, ref=gt, remap_measures=True)[0], 2)
        representation_accs_mesa13[representation] += acc
for k in representation_accs_mesa13:
    representation_accs_mesa13[k] = round(representation_accs_mesa13[k]/13,2)
```

```python
with open('valid_smr.pkl', 'rb') as f:
    valid_smr = pickle.load(f)
```

```python
representation_accs_smr = {k: 0 for k in ['frame_preds', 'frames', 'onset_preds', 'midi']}
for tag in tqdm(valid_smr):
    tag = str(tag)
    for representation in ['frame_preds', 'frames', 'onset_preds', 'midi']:
        aligned = smr.load_audio_feature_alignment(tag=tag, audio_feature=representation)
        gt = smr.load_smr_alignment(tag=tag)
        acc = round(evaluate(pred=aligned, ref=gt, remap_measures=True)[0], 2)
        representation_accs_smr[representation] += acc
for k in representation_accs_smr:
    representation_accs_smr[k] = round(representation_accs_smr[k]/len(valid_smr),2)
```
